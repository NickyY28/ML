```python
df = pd.DataFrame({
    'x' : list(range(-100,100,5))
})
df.head()
```

```python
plt.plot(df['x'])
```

```python
df['x2'] = df['x']**2
df['x3'] = df['x']**3
df['x4'] = df['x']**4
```

```python
df
```

```python
fig,(ax1,ax2,ax3,ax4) = plt.subplots(4,1, figsize=(8,12))
ax1.plot(df['x'], df['x'], color= 'blue')
ax2.plot(df['x'], df['x2'], color= 'red')
ax3.plot(df['x'], df['x3'], color= 'green')
ax4.plot(df['x'], df['x4'], color= 'purple')
ax1.legend()
ax2.legend()
ax3.legend()
ax4.legend()
plt.show()
```

```python
df = pd.read_csv("Position_Salaries.csv")
df
```

```python
import seaborn as sns
sns.regplot(df, x= 'Level', y= 'Salary', marker='o')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
X = df[['Level']]
y = df['Salary']
```

```python
pf = PolynomialFeatures(degree=4)
X_poly = pf.fit_transform(X)
```

```python
regressor=LinearRegression()
regressor.fit(X_poly,y)
```
